```python
#library for punctuation
import string
string.punctuation

#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
df['content_clean']= df['content'].apply(lambda x:remove_punctuation(x))
df.head()
```

```python
#lowering text
df['lower_text']= df['content_clean'].apply(lambda x: x.lower())
df.head()
```

```python
import re
#remove number
df['clear_number'] = df['lower_text'].apply(lambda x: re.sub(r'\d+', '', x))
df.head()
```

```python
#remove hastag
df['clear_hastag'] = df['clear_number'].apply(lambda x: re.sub(r'#\w+', '', x))
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

#remove stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

df['no_stopwords'] = df['clear_hastag'].apply(remove_stopwords)
df.head()
```

```python
import re
#remove emojis
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emojis'] = df['no_stopwords'].apply(remove_emojis)
df.head()
```

```python
#remove whitespace
def remove_whitespace(text):
    return ' '.join(text.split())

df['no_whitespace'] = df['no_emojis'].apply(remove_whitespace)
df.head()
```

```python
#check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#tokenized
def tokenization(text):
    tokens = re.split(r'\W+', text)
    return tokens
#applying function to the column
df['tokenized']= df['no_whitespace'].apply(lambda x: tokenization(x))
df.head()
```

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer

#defining the object for stemming
porter_stemmer = PorterStemmer()

#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text

df['stemming_text'] = df['tokenized'].apply(lambda x: stemming(x))
df.head()
```

```python
import spacy
import pandas as pd

#inisialisasi SpaCy dan muat model bahasa Inggris
nlp = spacy.load('en_core_web_sm')

#function for lemmatization using SpaCy
def lemmatization(tokens):
    doc = nlp(' '.join(tokens))
    lemmas = [token.lemma_ for token in doc]
    return lemmas

#function lemmatization from 'tokenized' to new column 'lemmatization_text'
df['lemmatization_text'] = df['tokenized'].apply(lambda x: lemmatization(x))
df.head()
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from transformers import pipeline
from scipy.special import softmax
```

```python
import pandas as pd
import numpy as np
import torch
from torch.nn.functional import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from torch.nn.functional import softmax
import pandas as pd

#function to get polarity scores using Roberta
def polarity_scores_roberta(stemming_text):
    #join the list of strings into a single string
    joined_text = ' '.join(stemming_text)
    encoded_text = tokenizer(joined_text, return_tensors='pt', max_length=512, truncation=True, padding=True)
    with torch.no_grad():
        output = model(**encoded_text)
    scores = output.logits[0]
    scores = softmax(scores, dim=0).numpy()
    scores_dict = {
        'roberta_neg': scores[0],
        'roberta_neu': scores[1],
        'roberta_pos': scores[2]
    }
    return scores_dict

#function to the DataFrame
df[['roberta_neg', 'roberta_neu', 'roberta_pos']] = df['stemming_text'].apply(
    lambda x: pd.Series(polarity_scores_roberta(x)))

#display the DataFrame
df.head()
```

```python
#function to assign sentiment label
def get_sentiment_label(scores):
    if scores['roberta_neg'] > scores['roberta_neu'] and scores['roberta_neg'] > scores['roberta_pos']:
        return 'negative'
    elif scores['roberta_neu'] > scores['roberta_neg'] and scores['roberta_neu'] > scores['roberta_pos']:
        return 'neutral'
    else:
        return 'positive'

#create the 'label_text' column
df['label_text1'] = df[['roberta_neg', 'roberta_neu', 'roberta_pos']].apply(get_sentiment_label, axis=1)

#display the dataFrame
df.head()
```
